```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
```

```python
recettes_echant = pd.read_csv("AUTRES/recettes-echant.data")
recettes_pays = pd.read_csv("AUTRES/recettes-pays.data")
```

```python
recettes_pays.head()
```

```python
recettes_echant.head()
```

```python
nb = recettes_echant.origin.value_counts()
nb = pd.DataFrame(nb)
nb["country"] = nb.index
nb = nb.reset_index(drop=True)
```

```python
ingredients = list(recettes_echant)
ingredients.remove('origin')
```

```python
nb_ingredients = [recettes_echant.groupby("origin")[col].apply(lambda x : sum(x)) for col in ingredients]
```

```python
df = pd.DataFrame(nb_ingredients)
df = df.T
df["country"] = df.index
df = df.reset_index(drop=True)
```

```python
df = pd.merge(df, nb, on="country", how='left')
df
```
