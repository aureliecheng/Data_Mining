## CLUSTERING

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
new_recettes_echant = recettes_echant.drop(["origin"], axis=1)
```

```python
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(new_recettes_echant)
    kmeanModel.fit(new_recettes_echant)
    distortions.append(sum(np.min(cdist(new_recettes_echant, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / new_recettes_echant.shape[0])
    
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(new_recettes_echant)

clusters = pd.DataFrame(y_kmeans)
```

```python
new_recettes_echant["clusters"] = clusters
recettes_echant["clusters"] = clusters
```

```python
v_clusters = recettes_echant[["origin", "clusters"]]
```

```python
clusters_0 = v_clusters[(v_clusters.clusters == 0)]
clusters_1 = v_clusters[(v_clusters.clusters == 1)]
clusters_2 = v_clusters[(v_clusters.clusters == 2)]
clusters_3 = v_clusters[(v_clusters.clusters == 3)]
# clusters_4 = v_clusters[(v_clusters.clusters == 4)]
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Histogram(x=clusters_0["origin"], name='cluster 0'))
fig.add_trace(go.Histogram(x=clusters_1["origin"], name='cluster 1'))
fig.add_trace(go.Histogram(x=clusters_2["origin"], name='cluster 2'))
fig.add_trace(go.Histogram(x=clusters_3["origin"], name='cluster 3'))
# fig.add_trace(go.Histogram(x=clusters_4["origin"], name='cluster 4'))
```

## CLASSIFICATION

```python
X = recettes_echant.drop(["origin"], axis=1)
y = recettes_echant[["origin"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(n_estimators=200, max_depth=12, random_state=0)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
print(classification_report(y_test, rfc_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

clf = AdaBoostClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)
clf_pred = clf.predict(X_test)

print(classification_report(y_test, clf_pred))
```
