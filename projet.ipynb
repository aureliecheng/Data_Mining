## Distribution 

```python
fig = px.histogram(recettes_echant, 
                   x="origin",
                  title="Nombre de recettes par origine").update_xaxes(categoryorder="total descending")
fig.show()
```

```python
recettes_echant.origin.unique().size
```

La distribution va nous permettre d'identifier les variables qui sont susceptibles de biaiser les résultats de notre modèle. En effet, la distribution est inégale avec une sur-représentation des recettes américaines (3 fois plus que les recettes italiennes) et une sous-représentation des recettes allemandes qui possède seulement 3 individus.
